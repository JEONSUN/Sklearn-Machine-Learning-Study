```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.random.randint(10,15,size = 15)
data1 = np.random.random(15)
data2 = data + data1

datay = np.random.random(15) + np.random.randint(0,10,size = 15)
```

```python
data = np.array([data2,datay])
```

```python
x = np.arange(-10,15)
y = np.arange(-7.5,10.0)
```

```python
# data
plt.scatter(data[0],data[1])
plt.axis([-5,20,-5,15.0])
plt.show()
```

## StandardScaler

- 각 특성의 평균을 0, 분산을 1로 변경하여 모든 특성이 같은 크기를 가지게 함.

- 특성의 최솟값과 최댓값 크기를 제한하지는 않음.

```python
from sklearn.preprocessing import StandardScaler
standard = StandardScaler()
data_standard = standard.fit_transform(data[0].reshape(-1,1))
data_standard1 = standard.fit_transform(data[1].reshape(-1,1))
```

```python
standard_data_scaler = np.array([data_standard, data_standard1])
plt.scatter(standard_data_scaler[0],standard_data_scaler[1])
plt.axis([-2,2,-2,2])
plt.show()
```

## RobustScaler

- (X - Q2) / (Q3-Q1)<br>
[Q2 = 중간값, Q3 = 3분위값, Q1 = 1분위값]
 
 
- 특성들이 같은 스케일을 갖게 된다는 측면에서는 StandardScaler와 비슷.


- 평균과 분산 대신 중간값과 사분위 값을 사용한다.<br>
-> 따라서 이상치의 영향을 받지 않는다.

```python
from sklearn.preprocessing import RobustScaler
robust = RobustScaler()
data_rob = robust.fit_transform(data[0].reshape(-1,1))
data_rob1 = robust.fit_transform(data[1].reshape(-1,1))
```

```python
rob_data_scaler = np.array([data_rob, data_rob1])
plt.scatter(rob_data_scaler[0],rob_data_scaler[1])
plt.axis([-2,2,-2,2])
plt.show()
```

## MinMaxScaler

- (X-Xmin)/(Xmax-Xmin)


- 모든 특성이 0과 1사이에 위치하도록 데이터를 변경


- 2차원의 데이터셋일 경우 모든 데이터가 x축의 0과 y축의 0과 1사이의 사각 영역에 담기게 됨.

```python
from sklearn.preprocessing import MinMaxScaler
minmax = MinMaxScaler()
data_minmax = minmax.fit_transform(data[0].reshape(-1,1))
data_minmax1 = minmax.fit_transform(data[1].reshape(-1,1))
```

```python
minmax_data_scaler = np.array([data_minmax, data_minmax1])
plt.scatter(minmax_data_scaler[0],minmax_data_scaler[1])
plt.axis([-2,2,-2,2])
plt.show()
```

```python
plt.figure(figsize = (10,8))

# 1번째 그레프
plt.subplot(131)
plt.scatter(standard_data_scaler[0],standard_data_scaler[1])
plt.axis([-2,2,-2,2])
plt.title('StandardScaler')

# 2번째 그래프
plt.subplot(132)
plt.scatter(rob_data_scaler[0],rob_data_scaler[1], c = 'g')
plt.axis([-2,2,-2,2])
plt.title('RobustScaler')

# 3번째 그래프
plt.subplot(133)
plt.scatter(minmax_data_scaler[0],minmax_data_scaler[1], c = 'r')
plt.axis([-2,2,-2,2])
plt.title('MinMaxScaler')

plt.show()
```
