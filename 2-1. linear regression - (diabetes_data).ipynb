```python
from sklearn.datasets import load_diabetes
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
diabetes = load_diabetes()
```

diabetes : 442 명의 당뇨병 환자 각각에 대해 10 개의 기준 변수, 연령, 성별, 체질량 지수, 평균 혈압 및 6 개의 혈청 측정 값

```python
diabetes
```

```python
print('data :', diabetes['data'])
```

```python
# 변수 명 확인
print('feature names :', diabetes['feature_names'])
```

```python
print('target :', diabetes['target'])
```

```python
# 훈련 데이터와 테스트 데이터 분리
from sklearn.model_selection import train_test_split
X_train, X_test, y_train,y_test = train_test_split(
diabetes['data'],diabetes['target'], random_state = 42)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
df_diabetes = pd.DataFrame(X_train, columns = diabetes['feature_names'])
df_diabetes.head()
```

### k-최근접 이웃 회귀

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
# 회귀모델 생성
# distance를 사용할 경우 가중평균 사용
# 기본값은 단순 평균인 uniform
regressor = KNeighborsRegressor(n_neighbors = 3, weights = 'distance')
```

```python
# 모델 훈련
regressor.fit(X_train,y_train)
```

```python
print("test 결과 예측 :" , regressor.predict(X_test))
```

```python
# 최근접 이웃 회귀는 보통 분류에 사용하므로 선형회귀 데이터셋에서는 r^2가 0.13으로 낮게 나옴
print("test R^2 : {:.2f}".format(regressor.score(X_test,y_test)))
```

### 선형 회귀

```python
from sklearn.linear_model import LinearRegression
```

```python
# 선형회귀 모델 훈련
#fit_intercept=True, : y절편을 계산할건지 여부
# normalize=True,  : 정규화를 시킬것인지
# n_jobs=None : 모델 계산에 사용할 작업 수

lr = LinearRegression(normalize = True).fit(X_train,y_train)
```

```python
# 회귀계수 계산값
print('lr.coef_:', lr.coef_) # 훈련 데이터에서 이끌어낸 속성은 끝에 _를 붙임

# y절편 계산값
print('lr.intercept_:', lr.intercept_)
```

```python
# 모델 정확도 계산 
print('train score: {:.2f}'.format(lr.score(X_train,y_train)))
print('test score : {:.2f}'.format(lr.score(X_test,y_test)))
```

```python
# 테스트의 예측값 생성
prediction = lr.predict(X_test)
print(prediction)
```

```python
# mean squared error 계산
from sklearn.metrics import mean_squared_error

print('mse : {:.2f}'.format(mean_squared_error(X_test[:,0],prediction)))
```

mse가 23793으로 예상보다 크게나옴.
