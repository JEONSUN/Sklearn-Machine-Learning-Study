## Decision Tree

결정트리는 의사결정 규칙을 나무구조로 나타내어 전체 데이터를 소집단으로 분류하거나 예측하는 분석기법을 의미한다.


### Ex) iris data

```python
# 예시 데이터셋 불러오기
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
```

```python
iris = load_iris()
x = iris.data[:,2:]
y = iris.target
```

```python
print(x.shape)
print(y.shape)
```

```python
# 결정 모델 훈련(트리 최대 깊이 2)
tree_clf = DecisionTreeClassifier(max_depth = 2)
```

```python
tree_clf.fit(x,y)
```

### export_graphviz()사용

```python
import os
PROJECT_ROOT_DIR = "D:\Portfolio\hands on machine learning dataset" ##파이썬 디렉토리 저장
CHAPTER_ID = "decision_trees"
IMAGE_PATH = os.path.join(PROJECT_ROOT_DIR, "images", CHAPTER_ID)

def image_path(fig_id, image_path_k2h=IMAGE_PATH):
    return os.path.join(image_path_k2h, fig_id) ##사진을 저장할 위치
```

```python
from sklearn.tree import export_graphviz
def export_graphviz_k2h(ML_algorithms, image_path_k2h=IMAGE_PATH):
    if not os.path.isdir(image_path_k2h):#github 코드로는 작동이 안되서
        os.makedirs(image_path_k2h) ##직접 image 폴더, decision_trees 폴더를 만듦
    export_graphviz(
    ML_algorithms,
    out_file=image_path("iris_tree.dot"), ##C:\Python\image\decisiontrees\iris_tree.dot
    feature_names=iris.feature_names[2:],
    class_names=iris.target_names,
    rounded=True, ##박스 생김새
    filled=True ##박스 생김새
        )

export_graphviz_k2h(tree_clf) ##dot 파일을 만들어줌
```

```python
import graphviz
with open("D:\Portfolio\hands on machine learning dataset\images\decision_trees\iris_tree.dot") as f:
    dot_graph = f.read()

dot = graphviz.Source(dot_graph)
dot.format = 'png'
dot.render(filename='iris_tree', directory='images/decision_trees', cleanup=True)
dot ##시스템 변수, PATH에 graphviz가 있어야함
```

나무에서 분할되는 부분을 **노드(node)**라고 하며 가장 처음 노드를 **뿌리마디(root node)**, 마지막 노드를 **끝마디(terminal node)**라고 한다.

결정 트리는 **root node(깊이가 0인 맨 꼭대기 노드)**에서 시작한다.

petal_width(꽃잎의 너비)가  0.8보다 같거나 작은지 검사하여 참이라면 왼쪽 **child node**이1)로 거짓이라면 오른쪽 노드로 간다.<br>
true로 갈 경우 더 이상 검사할 게 없으므로 **leaf node** 즉 자식노드를 가지지 않으므로 추가 검사를 하지 않는다.<br>
따라서 해당 노드에 있는 값들은 노드 예측값에 따라 setosa class로 예측한다.

꽃잎의 너비가 0.8보다 큰 경우 다시 한번 꽃잎의 너비가 1.75보다 작거나 같은지 검사하여 참이라면 왼쪽의 versicolor로, 아니라면 virginica로 분류, 예측한다.

### Decision tree의 단점

- 나무의 깊이가 깊어질수록 오버피팅이 심해진다.

해결 방안

- 나무의 최대 깊이(depth)를 제한<br>
- 자식 노드의 최소 샘플 수를 설정<br>


```python
import matplotlib.pyplot as plt
```

```python
X = np.random.sample(50)
Y = np.array([x**2 + np.random.normal(0, 0.05) for x in X])

plt.xlabel("x")
plt.ylabel("y")
plt.scatter(X, Y, color='b')
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor

plt.figure(figsize = (10,10))
# create a decision tree regressor
fit = DecisionTreeRegressor()

# and grow it (note that X must be reshaped)
fit.fit(np.reshape(X, (-1, 1)), Y);

# prepare test sample with "newaxis" trick
X_test = np.arange(0.0, 1.0, 0.01)[:, np.newaxis]
Y_test = fit.predict(X_test)

plt.scatter(X, Y, color='b')
plt.plot(X_test, Y_test);
```

## 규제 매개변수

결정 트리는 훈련데이터에 대한 제약 사항이 거의 없다. 따라서 제한을 두지 않는다면 모델이 훈련 데이터에 가깝게 맞추려고하기에 대부분 과대적합되기 쉽다.

훈련되기 이전의 파라미터 수가 결정되지 않은 모델을 **비파라미터 모델**이라고 한다.<br>
반대의 의미로는 선형 모델 같은 **파라미터 모델**이라고 하며 미리 정의된 모델 파라미터 수를 가지므로 자유도가 제한되고 과대적합될 위험이 줄어든다.
